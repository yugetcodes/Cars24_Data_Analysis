```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url = "https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&listingSource=ViewAllCars&storeCityId=2"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
links = soup.find_all('a')
for link in links:
    print(link.get('href'))
```

```python
cars = soup.find_all('div', {"class":"_2YB7p"})
```

```python
len(cars)
```

```python
car_details = []
for car in cars:
    try:
        name_string = car.find('h3', {'class': '_11dVb'}).get_text()
        
        # Extract year, company, and model using string splitting
        parts = name_string.split(' ', 2)
        year = parts[0]
        company = parts[1]
        model = parts[2]

        location = 'Delhi'
        price = car.find('strong', {'class':'_3RL-I'}).text
        
        # Find the <ul> element with the specified class
        ul_element = car.find('ul', {'class':'_3J2G-'})
        
        # Initialize variables
        fuel_type = None
        km_driven = None

        # Ensure ul_element is found before processing
        if ul_element:
            li_elements = ul_element.find_all('li')
            
            # Concatenate text from all <li> elements
            combined_text = ' '.join(li.get_text() for li in li_elements)

            # Use regular expression to match km driven and fuel type
            match = re.search(r'([\d,]+)\s*km\s*(\w+)', combined_text)
            if match:
                km_driven = match.group(1)
                fuel_type = match.group(2)

        car_details.append([name_string, location, model, price, fuel_type, company, year, km_driven])
    except Exception as e:
        print(f"Error processing car: {e}")

# Check the extracted data
for detail in car_details:
    print(detail)

# Save to Excel
df = pd.DataFrame(car_details, columns=['Name', 'Location', 'Model', 'Price', 'Fuel Type', 'Company', 'Year', 'KM Driven'])
df

```

```python
df.to_excel('car_details.xlsx', index=False)
```
